# 0x03 Overlap & Clustering
Explore infrastructure clustering using pandas and networkx.

```
import pandas as pd
import networkx as nx
import matplotlib.pyplot as plt

# Generate sample infrastructure data representing clustered SSH hashes
data = {
    'IP': ['198.51.100.1', '198.51.100.2', '203.0.113.5', '203.0.113.6', '198.51.100.1'],
    'Domain': ['c2-server1.evil.com', 'c2-server2.evil.com', 'phishing.net', 'c2-server1.evil.com', 'malware.com'],
    'SSH_Hash': ['hash_alpha', 'hash_beta', 'hash_alpha', 'hash_gamma', 'hash_alpha']
}
df = pd.DataFrame(data)
display(df)
```

```
# Create a graph to cluster by SSH Hash
G = nx.Graph()
for _, row in df.iterrows():
    G.add_edge(row['IP'], row['Domain'])
    if pd.notna(row['SSH_Hash']):
        G.add_edge(row['IP'], row['SSH_Hash'])

plt.figure(figsize=(10,8))
pos = nx.spring_layout(G)

# Color nodes by type
colors = []
for node in G.nodes():
    if 'hash' in node: colors.append('red')
    elif '.com' in node or '.net' in node: colors.append('lightblue')
    else: colors.append('lightgreen')

nx.draw(G, pos, with_labels=True, node_color=colors, edge_color='gray', node_size=1500, font_size=9)
plt.title("Infrastructure Clustering by SSH Hash (Red=Hash, Green=IP, Blue=Domain)")
plt.show()
```
