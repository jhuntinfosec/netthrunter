# 0x07 Graph-Based Hunting
Build and analyze threat infrastructure graphs.

```
import networkx as nx
import matplotlib.pyplot as plt

# Create a sample actor infrastructure graph
G = nx.Graph()
edges = [
    ("Threat_Actor_A", "198.51.100.4"),
    ("Threat_Actor_A", "198.51.100.5"),
    ("198.51.100.4", "phishing-login.com"),
    ("198.51.100.5", "malicious-update.net"),
    ("phishing-login.com", "ASN_12345 (Bulletproof Host)"),
    ("malicious-update.net", "ASN_12345 (Bulletproof Host)"),
    ("198.51.100.4", "JARM_Fingerprint_X"),
    ("198.51.100.99", "JARM_Fingerprint_X"), # Correlated IP
    ("198.51.100.99", "ASN_12345 (Bulletproof Host)")
]
G.add_edges_from(edges)

print("Nodes:", len(G.nodes))
print("Edges:", len(G.edges))
```

```
# Calculate centrality metrics
degree_centrality = nx.degree_centrality(G)
betweenness_centrality = nx.betweenness_centrality(G)

print("Top Nodes by Betweenness Centrality (Bottlenecks/Pivots):")
for node, score in sorted(betweenness_centrality.items(), key=lambda x: x[1], reverse=True)[:3]:
    print(f" - {node}: {score:.3f}")

# Visualize the graph
plt.figure(figsize=(12, 8))
pos = nx.spring_layout(G, seed=42)
nx.draw(G, pos, with_labels=True, node_size=2000, node_color='thistle', font_size=9, font_weight='bold', edge_color='gray')
plt.title("Threat Actor Infrastructure Pivot Graph")
plt.show()
```
